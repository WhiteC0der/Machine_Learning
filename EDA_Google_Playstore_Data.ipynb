```python
iteam = '19M'
iteam[-1]
```

```python
# 1 MB = 1024 KB
def size_process(item):
  if str(item)[-1] == 'M':
    res = float(str(item).replace('M', ''))
    return res*1028
  elif str(iteam)[-1] == 'k':
    res = float(str(item).replace('k', ''))
    return res
  else:
    return str(np.nan)
```

```python
df_copy = df.copy()
```

```python
df_copy['Size']=df_copy['Size'].apply(size_process)
```
